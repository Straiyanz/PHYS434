```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1) get points
mean = 3
std_dev = 0.01
N = 100_000

data_points = norm.rvs(loc=mean, scale=std_dev, size=N)
data_points
```

```python
# 2) plot points in histogram
x_range = (2.95, 3.05)
bins=50

plt.hist(data_points, bins=bins, density=True, range=x_range)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Gaussian Distribution')

plt.xticks(np.arange(x_range[0], x_range[1], 0.01))

plt.show()
```

```python
# 3) add PDF
plt.hist(data_points, bins=bins, density=True, range=x_range, label='Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency/PDF')
plt.title('Histogram and Gaussian PDF')

plt.xticks(np.arange(x_range[0], x_range[1], 0.01))

x_values = np.linspace(x_range[0], x_range[1], 100_000)
pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
plt.plot(x_values, pdf_values, color='red', label='PDF')
plt.legend()

plt.show()
```

```python
# 4) log scale
plt.hist(data_points, bins=bins, density=True, range=x_range, label='Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency/PDF (log)')
plt.title('Histogram and Gaussian PDF')

plt.xticks(np.arange(x_range[0], x_range[1], 0.01))

pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
plt.plot(x_values, pdf_values, color='red', label='PDF')
plt.yscale('log')
plt.legend()

plt.show()
```
